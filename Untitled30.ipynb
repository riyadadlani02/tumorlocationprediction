```python
import os
import numpy as np
from PIL import Image, UnidentifiedImageError
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

extracted_path = 'remaining_images'  # Update this to your actual path
your_target_size = (64, 64)  # Replace with your target size
your_feature_dim = 3  # Replace with your feature dimensions (e.g., 3 for RGB)
your_epochs = 10  # Replace with the number of epochs

image_files = []
labels = []

for root, dirs, files in os.walk(extracted_path):
    for file in files:
        if file.endswith(".jpg"):
            image_path = os.path.join(root, file)
            try:
                img = Image.open(image_path).convert("RGB").resize(your_target_size)
                image_files.append(np.array(img))
                labels.append(os.path.basename(root))
            except (IOError, OSError, UnidentifiedImageError):
                print(f"Skipping invalid image: {image_path}")

# Convert to numpy arrays
X = np.array(image_files)
y = np.array(labels)

# Encode labels using LabelEncoder
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Normalize pixel values to be between 0 and 1
X_train, X_test = X_train / 255.0, X_test / 255.0

# Define the CNN model
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=your_target_size + (your_feature_dim,)))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(np.max(y_encoded) + 1, activation='softmax'))  # Adjust the number of neurons in the output layer

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=your_epochs, validation_data=(X_test, y_test))

# Evaluate the model
accuracy = model.evaluate(X_test, y_test)[1]
print(f"Model Accuracy: {accuracy * 100:.2f}%")
```

```python
import zipfile
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv3D, MaxPooling3D, Flatten, Dense
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing import image
from PIL import Image

# Assuming you have your zip file uploaded to Colab
zip_path = 'images.zip'
extracted_folder = 'extracted_images'

# Extract the contents of the zip file
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extracted_folder)

# Dynamically assign class labels based on subfolder names
class_names = sorted(os.listdir(extracted_folder))
class_labels = {class_name: idx for idx, class_name in enumerate(class_names)}
num_classes = len(class_names)

# Load images and labels
data = []
labels = []
for class_name in class_names:
    class_path = os.path.join(extracted_folder, class_name)
    class_label = class_labels[class_name]

    for img_file in os.listdir(class_path):
        img_path = os.path.join(class_path, img_file)

        try:
            img = image.load_img(img_path, target_size=(64, 64, 64), color_mode='grayscale')
            img_array = image.img_to_array(img)
            data.append(img_array)
            labels.append(class_label)
        except (OSError, UnidentifiedImageError, Image.DecompressionBombError):
            print(f"Skipping invalid image: {img_path}")

X = np.array(data)
y = np.array(labels)

# Perform one-hot encoding on labels
y_encoded = to_categorical(y, num_classes=num_classes)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# ... (rest of the code remains the same)
```
